## If-Statements with User-Defined Functions

```python
# Create User-Defined Functions indicating low, mid, and high-range products in terms of item price 
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15: 
        return 'High-range product'
    else: return 'not enough data' 
```

```python
# Now that the price_label has been defined, use it on dataframe 
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Frequency check of new column 
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

```python
# If Statements using loc() function 
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## If Statements with For-Loops

```python
# Determine which day the most orders take place 
# Saturday is the busiest day, Wednesday is the least busy day 
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a new column
result = []
for value in ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
ords_prods_merge['busiest_day'] = result 
```

```python
# Create a statement containing busiest days and least busiest days 
result = []
for value in ords_prods_merge['orders_day_of_week']:
    if value == 0 or value == 1:
        result.append('Busiest days')
    elif value == 3 or value == 4:
        result.append('Slowest days')
    else:
        result.append('Regularly busy')
```

```python
# Verifying output 
result
```

```python
# Rename busiest day to busiest days
ords_prods_merge.rename(columns = {'busiest_day' : 'busiest_days'}, inplace = True)

# input results into busiest_days column
ords_prods_merge['busiest_days'] = result 
```

```python
# Verify results 
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

When combining the busiest days, the value comes very close to the regularly busy days as opposed to before there was a significant gap. 

```python
# Determine the traffic for each hour of the day
ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

10AM and 11AM are the busiest hours of the day.

```python
# Divide the hours into thirds to demonstrate which hours have the most orders, average orders, and fewest orders
result = []
for value in ords_prods_merge['order_hour_of_day']:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result.append('Most orders')
    elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
        result.append('Average orders')
    else:
        result.append('Fewest orders')
```

```python
result
```

```python
# save new results to dataframe 
ords_prods_merge['busiest_period_of_day'] = result 
```

```python
# verify results 
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
